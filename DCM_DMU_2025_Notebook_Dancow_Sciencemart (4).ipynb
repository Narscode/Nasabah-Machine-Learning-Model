```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Agregasi data
pendidikan_agg = df_training.groupby('pendidikan')['berlangganan_deposito'].agg(['count', 'sum', 'mean']).reset_index()
pendidikan_agg.rename(columns={
    'count': 'jumlah_nasabah',
    'sum': 'jumlah_berlangganan',
    'mean': 'rasio_berlangganan'
}, inplace=True)

# Urutkan agar lebih rapi saat ditampilkan
pendidikan_agg.sort_values(by='rasio_berlangganan', ascending=False, inplace=True)

# Plot 1: Jumlah nasabah dan yang berlangganan
plt.figure(figsize=(14, 6))
sns.barplot(data=pendidikan_agg, x='pendidikan', y='jumlah_nasabah', color='skyblue', label='Jumlah Nasabah')
sns.barplot(data=pendidikan_agg, x='pendidikan', y='jumlah_berlangganan', color='steelblue', label='Berlangganan')
plt.title('Jumlah Nasabah dan yang Berlangganan per Pendidikan')
plt.ylabel('Jumlah Orang')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot 2: Rasio keberhasilan (mean)
plt.figure(figsize=(12, 5))
sns.barplot(data=pendidikan_agg, x='pendidikan', y='rasio_berlangganan', palette='viridis')
plt.title('Rasio Keberhasilan Berlangganan per Pendidikan')
plt.ylabel('Rasio Berlangganan')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# --- Imports ---
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.impute import KNNImputer, SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from imblearn.over_sampling import SMOTE
from sklearn.metrics import roc_auc_score
from collections import Counter
import lightgbm as lgb
from lightgbm.basic import LightGBMError

# --- Data Loading ---
df_training = pd.read_csv('training_dataset.csv')
df_validation = pd.read_csv('validation_set.csv')

# --- Drop 'customer_number' if exists ---
df_training = df_training.drop(columns=['customer_number'], errors='ignore')
df_validation = df_validation.drop(columns=['customer_number'], errors='ignore')

# --- Replace categorical yes/no/unknown ---
cols_binary = ['pinjaman_rumah', 'gagal_bayar_sebelumnya', 'pinjaman_pribadi']
df_training[cols_binary] = df_training[cols_binary].replace({'yes': 1, 'no': 0, 'unknown': np.nan})
df_validation[cols_binary] = df_validation[cols_binary].replace({'yes': 1, 'no': 0, 'unknown': np.nan})

# --- Replace 999 with NaN ---
df_training['hari_sejak_kontak_sebelumnya'] = df_training['hari_sejak_kontak_sebelumnya'].replace(999, np.nan)
df_validation['hari_sejak_kontak_sebelumnya'] = df_validation['hari_sejak_kontak_sebelumnya'].replace(999, np.nan)

# --- Convert month and day to numeric ---
month_map = {'jan':1,'feb':2,'mar':3,'apr':4,'may':5,'jun':6,'jul':7,'aug':8,'sep':9,'oct':10,'nov':11,'dec':12}
day_map = {'mon':1,'tue':2,'wed':3,'thu':4,'fri':5,'sat':6,'sun':7}

df_training['bulan_kontak_terakhir'] = df_training['bulan_kontak_terakhir'].astype(str).str.strip().str.lower().replace(month_map)
df_validation['bulan_kontak_terakhir'] = df_validation['bulan_kontak_terakhir'].astype(str).str.strip().str.lower().replace(month_map)

df_training['hari_kontak_terakhir'] = df_training['hari_kontak_terakhir'].astype(str).str.strip().str.lower().replace(day_map)
df_validation['hari_kontak_terakhir'] = df_validation['hari_kontak_terakhir'].astype(str).str.strip().str.lower().replace(day_map)

# --- Imputation ---
numeric_cols = df_training.select_dtypes(include=np.number).columns
df_training[numeric_cols] = KNNImputer(n_neighbors=5).fit_transform(df_training[numeric_cols])
df_validation[numeric_cols.intersection(df_validation.columns)] = KNNImputer(n_neighbors=5).fit_transform(df_validation[numeric_cols.intersection(df_validation.columns)])

cat_cols = df_training.select_dtypes(include=['object', 'category']).columns
df_training[cat_cols] = SimpleImputer(strategy="constant", fill_value="missing").fit_transform(df_training[cat_cols])
df_validation[cat_cols.intersection(df_validation.columns)] = SimpleImputer(strategy="constant", fill_value="missing").fit_transform(df_validation[cat_cols.intersection(df_validation.columns)])

# --- Separate target ---
target = 'berlangganan_deposito'
X = df_training.drop(columns=target)
y = df_training[target]

# --- Encoding ---
categorical_cols = X.select_dtypes(include=['object', 'category']).columns
numerical_cols = X.select_dtypes(include=np.number).columns

preprocessor = ColumnTransformer([
    ('num', 'passthrough', numerical_cols),
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
])

X_encoded = preprocessor.fit_transform(X)

# --- SMOTE ---
print("Before SMOTE:", Counter(y))
X_res, y_res = SMOTE(random_state=42).fit_resample(X_encoded, y)
print("After SMOTE:", Counter(y_res))

# --- Split train/val ---
X_train, X_val, y_train, y_val = train_test_split(X_res, y_res, test_size=0.2, random_state=42, stratify=y_res)

# --- LightGBM Training ---
params = {
    'objective': 'binary',
    'metric': 'auc',
    'boosting_type': 'gbdt',
    'n_estimators': 1000,
    'learning_rate': 0.05,
    'feature_fraction': 0.9,
    'bagging_fraction': 0.8,
    'bagging_freq': 5,
    'verbose': -1,
    'n_jobs': -1,
    'seed': 42
}

model = lgb.LGBMClassifier(**params)

try:
    model.fit(X_train, y_train, eval_set=[(X_val, y_val)], eval_metric='auc', callbacks=[lgb.early_stopping(100, verbose=False)])
    y_proba = model.predict_proba(X_val)[:, 1]
    auc = roc_auc_score(y_val, y_proba)
    print(f"AUC Score on Validation Set: {auc:.4f}")
except LightGBMError as e:
    print("LightGBM Error:", e)
except Exception as e:
    print("Unexpected error:", e)

# --- Submission CSV creation ---

# Reload df_validation to get customer_number (if dropped previously)
df_val_original = pd.read_csv('validation_set.csv')
df_val_customer_number = df_val_original['customer_number']

# Re-apply preprocessing on validation data for prediction
df_validation[cols_binary] = df_validation[cols_binary].replace({'yes': 1, 'no': 0, 'unknown': np.nan})
df_validation['hari_sejak_kontak_sebelumnya'] = df_validation['hari_sejak_kontak_sebelumnya'].replace(999, np.nan)
df_validation['bulan_kontak_terakhir'] = df_validation['bulan_kontak_terakhir'].astype(str).str.strip().str.lower().replace(month_map)
df_validation['hari_kontak_terakhir'] = df_validation['hari_kontak_terakhir'].astype(str).str.strip().str.lower().replace(day_map)
df_validation[numeric_cols.intersection(df_validation.columns)] = KNNImputer(n_neighbors=5).fit_transform(df_validation[numeric_cols.intersection(df_validation.columns)])
df_validation[cat_cols.intersection(df_validation.columns)] = SimpleImputer(strategy="constant", fill_value="missing").fit_transform(df_validation[cat_cols.intersection(df_validation.columns)])

X_val_processed = preprocessor.transform(df_validation)

# Predict probability
y_val_proba = model.predict_proba(X_val_processed)[:, 1]

# Create submission dataframe
df_submission = pd.DataFrame({
    'customer_number': df_val_customer_number,
    'berlangganan_deposito': y_val_proba
})

# Save submission file
df_submission.to_csv('submission.csv', index=False)
print("Submission file 'submission.csv' berhasil dibuat.")
```
